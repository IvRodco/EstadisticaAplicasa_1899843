```python
#primero importamos las librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#abrimos la base de datos
pelis= pd. read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/peliculas.csv',index_col=0)
pelis
```

```python
## a)

# Cuantas peliculas hay?

Peliculas=pelis.groupby('title')['release_year'].count()
Contar_Peliculas=Peliculas.count()

# Cuantos generos hay?

Generos=pelis.groupby('category')['release_year'].count()
Contar_Generos=Generos.count()

# Promedio de duracion de las peliculas

Duracion=pelis['length'].mean()



print('El total de peliculas en la base de datos es:',Contar_Peliculas)
print('El total de generos en la base de datos es:',Contar_Generos)
print('La duracion promedio en minutos por pelicula es:',Duracion)
```

```python
## b)

Peliculas_dormir= pd.DataFrame(pelis,columns=['category','length'])
Peliculas_dormir=Peliculas_dormir[Peliculas_dormir['category'] == 'Documentary']
Peliculas_dormir=Peliculas_dormir[Peliculas_dormir['length'] >= 120]
Peliculas_dormir
```

```python
## c)
#Cuantas peliculas hay de caga genero?
TG=pelis.groupby('category')['category'].count()
TG
```

```python
## d)
#Cual es la duracion promedio de cada genero?

Promedio_Genero=pelis.groupby('category')['length']
Promedio_Genero.mean()
```

```python
## e)

condiciones=[pelis['length']<=90,
             pelis['length']<150,
             pelis['length']>=150]
valor_condicion=['Corta','Media','Larga']
pelis['Tipo_duracion']= np.select(condiciones,valor_condicion,default='Dato invalido')
pelis
```

```python
#traemos la base de datos
Poblacion=pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/poblacion_Asgard.csv',index_col=0)
Poblacion=Poblacion.rename(columns={'PaÃ­s':'Pais'})
Poblacion=Poblacion.rename(columns={'PoblaciÃ³n':'Poblacion'})
Poblacion
```

```python
## a)

Poblacion=Poblacion.sort_values(['Estado','Municipio'])

Poblacion
```

```python
## b)

#Cuantos estados hay?

EstadosTOT=Poblacion.groupby('Estado')['Estado'].count()
EstadosTOT=EstadosTOT.count()

#Cuantos municipios hay?
MunicipiosTOT=Poblacion.groupby('Municipio')['Municipio'].count()
MunicipiosTOT=MunicipiosTOT.count()

##
print('El total de estados es de:',EstadosTOT)
print('El total de municipios es de:',MunicipiosTOT)
```

```python
#Cuantos Municipios por cada estado hay?

Poblacion.groupby('Estado')['Municipio'].count()
```

```python
## c)

#Cuanta poblacion hay en total?
PoblacionTOT=Poblacion['Poblacion'].sum()
print('La poblacion total es de:',PoblacionTOT)
```

```python
#Cual es la poblacion por estado
Poblacion_Estado=Poblacion.groupby('Estado')['Poblacion'].sum()
Poblacion_Estado
```

```python
# d)
Poblacion_Municipio=Poblacion.groupby('Municipio')['Poblacion'].sum()
Poblacion_MunicipioTOT=Poblacion_Municipio.median()
print('La mediana de la poblacion por Municipio es:',Poblacion_MunicipioTOT)
```

```python
Pobl_Mun= pd.DataFrame(Poblacion,columns=['Municipio','Poblacion'])
Pobl_Mun['Tipo de poblacion']=np.where(Pobl_Mun['Poblacion']>=Poblacion_MunicipioTOT,'Muy poblado','Poco poblado')
Pobl_Mun=Pobl_Mun.sort_values('Poblacion')
Pobl_Mun
```

```python
# e)

# Municipios menos poblado
MENOSPOB=Pobl_Mun.head(5)
MENOSPOB
```

```python
# MUnicipios mas poblados
MASPOB=Pobl_Mun.tail(5)
MASPOB
```
